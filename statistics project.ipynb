# Importing Important Library

```python
import pandas as pd
import numpy as np
```

# Reading the Dataset

```python
df=pd.read_csv ("Laliga.csv")
df
```

```python
df.info()
```

# 1.Replacing dashes with 0 to make sure we can perform arithmetic    operations on the data


```python
df= df.replace('-', '0')
df
```

# Changing Datatype from string to numeric

```python
df['Points']=pd.to_numeric(df['Points'])
df['GoalsFor']=pd.to_numeric(df['GoalsFor'])
df['GoalsAgainst']=pd.to_numeric(df['GoalsAgainst'])
df['GamesWon']=pd.to_numeric(df['GamesWon'])
df['GamesPlayed']=pd.to_numeric(df['GamesPlayed'])
df.info()
```

# Describing the data set

```python
df.describe()
```

# 2.Printing all the teams which have started playing between 1930-1980

```python
Debut_year=df[df['Debut'].astype(str).str[:4].astype(int).between(1930,1980)]
Debut_year[['Team','Debut']]
```

# 3.Printing the list of teams which came Top 5 in terms of points

```python
top5=df.sort_values(by='Points',ascending = False)
top5[['Team','Points']].head(5)
```

# 4.Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences

```python
def Goal_diff_count():
    df[['Goal_diff_count']]=df['GoalsFor']-df['GoalsAgainst']
    return df[['Team','Goal_diff_count']].sort_values(by='Goal_diff_count',ascending = False)
Goal_diff_count()
```

```python
Goal_diff_count().head(1) #for maximum goal difference
```

```python
Goal_diff_count().tail(1)  #for minimum goal difference
```

# 5.Createing a new column with name “Winning Percent” and appending it to the data set

```python
df['Winning Percent']=df['GamesWon']/df['GamesPlayed']*100
df
df.sort_values(by='Winning Percent',ascending = False)
df.head(5)
```

# 6.Group teams based on their “Best position” and print the sum of their points for allpositions

```python
group=df[['BestPosition','Points']].groupby('BestPosition')
group.sum()
```
